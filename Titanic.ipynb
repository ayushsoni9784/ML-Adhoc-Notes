```python
df=pd.read_csv("titanic.csv")
```

```python
x=df.iloc[:,0:].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
# converting sex column into int 
x[:,3]=le.fit_transform(x[:,3])
```

```python
x[:,2]=le.fit_transform(x[:,2])
```

```python
x
```

```python
# now using OneHotEncoder on survived column 
from sklearn.preprocessing import OneHotEncoder
```

```python
firstcl=OneHotEncoder(categorical_features=[0])
```

```python
x=firstcl.fit_transform(x).toarray() 
```

```python
x.astype(int)
```

```python
features=x
```

```python
label=x[:,0]
```

```python
label1=np.where(label==1,'dead',label)
```

```python
label1=np.where(label1==0,'survived',label)
```

```python
label=np.where(label==1,'dead',label)
label
```

```python
l=label[:,]
```

```python
l=np.where(l=='0.0','survived',l)
```

```python
l
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,train_test,label_train,label_test=train_test_split(features,l,test_size=0.2) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kclf=KNeighborsClassifier(n_neighbors=5) 
```

```python
ktrained=kclf.fit(train_data,label_train )
```

```python
predict_output=ktrained.predict(train_test)
```

```python
predict_output
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(label_test,predict_output)
```
